```python
# Q1_0118. english와 mathematices 2차원 데이터간의 관계를 시각화해보세요

# 히스토그램
# 박스플랏
# scatter plot(regplot, implot)
# jointplot
# heatmap

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  


# CSV 파일 읽기
df = pd.read_csv('ch2_scores_em.csv')

# 히스토그램
df['english'].plot(kind='hist', alpha=0.5)
df['mathematics'].plot(kind='hist', alpha=0.5)
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Histogram of English and Mathematics Scores')
plt.show()

# 박스플랏
df[['english', 'mathematics']].plot(kind='box')
plt.title('Box Plot of English and Mathematics Scores')
plt.show()

# Scatter plot(regplot, lmplot)
sns.regplot(x='english', y='mathematics', data=df)
plt.title('Scatter Plot of English vs Mathematics (Regplot)')
plt.show()

sns.lmplot(x='english', y='mathematics', data=df)
plt.title('Scatter Plot of English vs Mathematics (Lmplot)')
plt.show()

# Jointplot
sns.jointplot(x='english', y='mathematics', data=df, kind='reg')
plt.show()

# Heatmap
corr = df[['english', 'mathematics']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
# Q2_0118. 주어진 숫자 배열 [4,8,15,16,23,42] 에 대해 평균과 중앙값을 계산한 후 배열에 두 개의 극단값을 추가한 후(예:-100, 100), 다시 평균과 중앙값을 계산하고 비교해보세요


import numpy as np

numbers = np.array([4, 8, 15, 16, 23, 42])

# 평균과 중앙값 계산
mean_original = np.mean(numbers)
median_original = np.median(numbers)

# 극단값을 추가한 후의 배열
numbers_extreme = np.append(numbers, [-100, 100])

# 극단값을 추가한 후의 평균과 중앙값 계산
mean_extreme = np.mean(numbers_extreme)
median_extreme = np.median(numbers_extreme)

print("원래 평균:", mean_original)
print("원래 중앙값:", median_original)
print("극단값 추가 평균:", mean_extreme)
print("극단값 추가 중앙값:", median_extreme)

```

```python
# Q3_0118. 아래 두개의 데이터로 공부산과 상관계수를 계산하세요

# data1 = [65,72,78,65,72,70,65,68]
# data2 = [72,69,74,66,70,75,65,72]

# 데이터 정의
data1 = [65, 72, 78, 65, 72, 70, 65, 68]
data2 = [72, 69, 74, 66, 70, 75, 65, 72]

# NumPy 배열로 변환
data1_np = np.array(data1)
data2_np = np.array(data2)

# 공분산 행렬
covariance = np.cov(data1_np, data2_np)

# 상관계수 행렬
correlation = np.corrcoef(data1_np, data2_np)

# 공분산과 상관계수만 별도로 추출
covariance_value = covariance[0, 1]  # data1과 data2의 공분산
correlation_value = correlation[0, 1]  # data1과 data2의 상관계수

# 추출된 값 출력
print("공분산:", covariance_value)
print("상관계수:", correlation_value)

```

```python
#Q5_0118. 주어진 데이터 배열의 제1사분위수, 중앙값, 제3사분위수, 사분위수를 계산한 후 데이터의 10번째 백분위수와 90번째 백분위수를 계산하여 출력하세요.

# 데이터 정의
data = np.arange(1000)

# 제1사분위수, 중앙값(제2사분위수), 제3사분위수 계산
Q1 = np.percentile(data, 25)  # 제1사분위수 (25번째 백분위수)
median = np.median(data)      # 중앙값 (50번째 백분위수)
Q3 = np.percentile(data, 75)  # 제3사분위수 (75번째 백분위수)

# 사분위수 범위(IQR) 계산
IQR = Q3 - Q1

# 10번째 백분위수와 90번째 백분위수 계산
P10 = np.percentile(data, 10)  # 10번째 백분위수
P90 = np.percentile(data, 90)  # 90번째 백분위수

print('제1사분위수:', Q1)
print('중앙값:', median)
print('제3사분위수:', Q3)
print('사분위수 범위(IQR):', IQR)
print('10번째 백분위수:', P10)
print('90번째 백분위수:', P90)


```

```python
#Q6_0118. 아래 데이터를 0~1 범위로 정규화 하여 출력하세요

from sklearn.preprocessing import MinMaxScaler

# 데이터 정의
data = np.array([[1], [2], [3], [4], [5]])

# MinMaxScaler를 사용하여 0~1 범위로 정규화
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(data)

# 결과 출력
print('정규화된 데이터:')
print(normalized_data)

```

```python
# Q7_0118. Z-점수 정규화를 사용하여 주어진 데이터 배열을 표준화하세요. (평균이 0이고 분산이 1이 되도록 특성의 크기를 조정)

from sklearn.preprocessing import StandardScaler

# 데이터 정의
data = np.array([[1], [2], [3], [4], [5]])

# StandardScaler를 사용하여 Z-점수 정규화 (평균 0, 분산 1)
scaler = StandardScaler()
standardized_data = scaler.fit_transform(data)

# 결과 출력
print('표준화된 데이터:')
print(standardized_data)

```
