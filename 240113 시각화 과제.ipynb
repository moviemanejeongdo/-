```python
import pandas as pd
import numpy as np

# 주가 데이터 CSV 파일 읽기
df = pd.read_csv('주가데이터.csv')

# 'Date' 칼럼이 존재하는지 확인
if 'Date' in df.columns:
    # 'Date' 칼럼을 datetime 타입으로 변환
    df['NDate'] = pd.to_datetime(df['Date'])

    # 'NDate'를 인덱스로 설정하고 'Date', 'Volume' 칼럼 제거
    df1 = df.set_index('NDate')
    df1.drop(['Date', 'Volume'], axis=1, inplace=True)

    # 결과 출력
    print(df1.head())
else:
    print("DataFrame에 'Date' 칼럼이 존재하지 않습니다.")
```

```python
import matplotlib.pyplot as plt

# df1 데이터프레임의 데이터를 선 그래프로 시각화
df1.plot()
```

```python
# Q1_0115. df1 Close 칼럼으로 다음을 수행하세요

# -NDate 별 막대그래프
# -6월만을 추출하여 Close 칼럼으로 일별 막대 그래프를 표현
# -히스토그램

import matplotlib.pyplot as plt

# NDate 별 막대그래프 (Close 칼럼)
df1['Close'].plot(kind='bar')
plt.title('NDate / Close')
plt.xlabel('NDate')
plt.ylabel('Close')
plt.show()

# 6월 데이터 추출 후 막대그래프
df_june = df1[df1.index.month == 6]
df_june['Close'].plot(kind='bar')
plt.title('June / Close')
plt.xlabel('day')
plt.ylabel('Close')
plt.show()

# 히스토그램
df1['Close'].plot(kind='hist', bins=20)
plt.title('Close histogram')
plt.xlabel('Close')
plt.ylabel('frequency')
plt.show()



```

```python
# Q2_0115. df에서 volume 칼럼의 분포를 탐색하는 히스토그램을 표현하세요.

# 파일 경로 설정 (여기서는 예시로 동일한 디렉토리 내 '주가데이터.csv' 파일을 사용한다고 가정)
file_path = '주가데이터.csv'

# CSV 파일에서 데이터 불러오기
df = pd.read_csv(file_path)

# 불러온 데이터 확인
print(df.head())

# Volume 칼럼에 대한 히스토그램 생성
plt.figure(figsize=(10, 6))     #가로 10인치 세로 6인치
plt.hist(df['Volume'], bins=10, color='blue', alpha=0.7)   #10개의 막대로
plt.title('Volume Distribution')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Q3_0115. df Close 칼럼과 Volume 칼럼으로 산점도를 표현하세요

# Close 칼럼과 Volume 칼럼으로 산점도 그리기
plt.figure(figsize=(10, 6))
plt.scatter(df['Close'], df['Volume'], color='green', alpha=0.5)
plt.title('Close vs Volume Scatter Plot')
plt.xlabel('Close')
plt.ylabel('Volume')
plt.grid(True)
plt.show()
```

```python
# Q4_0115. df Close 칼럼의 Box Plot을 표현하세요

plt.figure(figsize=(10, 6))
plt.boxplot(df['Close'])
plt.title('Close Column Box Plot')
plt.ylabel('Close')
plt.grid(True)
plt.show()
```
