```python
# Q3_0115. df Close 칼럼과 Volume 칼럼으로 산점도를 표현하세요

# Close 칼럼과 Volume 칼럼으로 산점도 그리기
plt.figure(figsize=(10, 6))
plt.scatter(df['Close'], df['Volume'], color='green', alpha=0.5)
plt.title('Close vs Volume Scatter Plot')
plt.xlabel('Close')
plt.ylabel('Volume')
plt.grid(True)
plt.show()
```

plt.scatter(df['Close'], df['Volume'], color='green', alpha=0.5) 코드에서 alpha는 그래프의 투명도를 설정하는 매개변수입니다. alpha 값은 0과 1 사이의 숫자로 설정할 수 있으며, 이 값이 낮을수록 그래프의 점들이 더 투명하게 표시됩니다.

alpha=1의 경우, 점들은 완전히 불투명합니다.
alpha=0의 경우, 점들은 완전히 투명하며 보이지 않습니다.
alpha=0.5의 경우, 점들은 중간 정도의 투명도를 갖게 됩니다.
이 매개변수는 데이터 포인트가 많을 때 유용합니다. 많은 수의 점들이 겹쳐 있을 때, alpha를 조절하면 데이터 포인트들 사이의 밀도를 더 쉽게 구별할 수 있게 해줍니다.

박스 플롯(Box Plot)은 데이터의 분포와 중앙값, 이상치를 시각적으로 나타내는데 사용되는 그래프입니다. 통계적 분석에서 매우 유용하며, 데이터의 범위와 중앙값, 분산 등을 한눈에 파악할 수 있습니다.

박스 플롯의 주요 구성 요소는 다음과 같습니다:

1. **중앙값 (Median)**: 데이터의 중간값을 나타냅니다. 박스 플롯의 중간에 위치한 선으로 표시됩니다.

2. **사분위수 (Quartiles)**: 
   - **제1사분위수 (Q1)**: 데이터의 하위 25%에 해당하는 값입니다.
   - **제3사분위수 (Q3)**: 데이터의 상위 25%에 해당하는 값입니다.
   - 박스의 하단과 상단은 각각 Q1과 Q3을 나타냅니다.

3. **IQR (Interquartile Range)**: Q3와 Q1의 차이로, 데이터의 중간 50% 범위를 나타냅니다.

4. **휘스커 (Whiskers)**: 박스의 상단과 하단에서 연장된 선으로, 일반적으로 Q1과 Q3에서 1.5 * IQR 범위 안에 있는 최소값과 최대값까지 표시됩니다.

5. **이상치 (Outliers)**: 휘스커의 바깥에 위치한 개별 점들로, 일반적인 범위를 벗어난 데이터 포인트를 나타냅니다.


박스 플롯은 데이터의 분포를 빠르게 파악하고 이상치를 식별하는 데 유용합니다. 특히 여러 그룹의 데이터를 비교할 때 효과적입니다.

박스 플롯에서 이상치를 판단하는 기준은 일반적으로 사분위수 범위(IQR)를 기반으로 합니다. 이상치는 데이터 중에서 대부분의 값들과 크게 다른 값을 가진 데이터 포인트를 의미합니다. 이상치를 구하는 구체적인 방법은 다음과 같습니다:

1. **사분위수 계산**: 먼저 데이터를 정렬하고 제1사분위수(Q1)와 제3사분위수(Q3)를 계산합니다. Q1은 데이터의 하위 25%, Q3는 상위 25%에 해당하는 값입니다.

2. **IQR 계산**: IQR은 Q3와 Q1의 차이로, 데이터의 중간 50% 범위를 나타냅니다. IQR = Q3 - Q1

3. **이상치 판단 범위 설정**: 이상치는 일반적으로 Q1에서 1.5 * IQR보다 낮거나, Q3에서 1.5 * IQR보다 높은 값을 가진 데이터 포인트로 정의됩니다. 즉, 이상치는 다음 범위 밖에 있는 값들입니다:
   - 아래쪽 경계: Q1 - 1.5 * IQR
   - 위쪽 경계: Q3 + 1.5 * IQR

이러한 방법으로 박스 플롯에서 이상치를 식별할 수 있으며, 이상치는 보통 박스 플롯 상에 점으로 표시됩니다. 이상치의 식별은 데이터 분석에서 중요한 부분이며, 데이터의 특성과 분포를 이해하는 데 도움을 줍니다.
